```python
from tools.vizualization import plot_length_distribution
```

```python
with open("../data/zh_sent_dataset.tsv") as f:
    sentences = f.readlines()
```

```python
# Compute max and average length (in characters) of sentences
sentence_lengths = [len(sentence.strip()) for sentence in sentences]
max_len = max(sentence_lengths)
avg_len = sum(sentence_lengths) / len(sentence_lengths)

print(f"Maximum sentence length: {max_len}")
print(f"Average sentence length: {avg_len:.2f}")
```

```python
plot_length_distribution(sentence_lengths, title="Sentence Length Distribution (Characters)")
```
